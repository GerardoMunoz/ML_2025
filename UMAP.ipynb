```python
import numpy as np
import umap
import matplotlib.pyplot as plt

# Step 1: Define the 3D parametric spring function
t = np.linspace(0, 1, 1000)  # 1000 points along the helix
x = np.cos(2 * np.pi * t)    # Circular motion in x
y = np.sin(2 * np.pi * t)    # Circular motion in y
z = t                        # Linear motion in z
X = np.vstack((x, y, z)).T   # 3D data points

# Step 2: Apply UMAP (3D → 2D)
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2)
X_umap = reducer.fit_transform(X)

# Step 3: Plot the 3D Spring
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x, y, z, c=t, cmap="Spectral", s=5)
ax1.set_title("Original 3D Spring (Helix)")

# Step 4: Plot the UMAP 2D Projection
ax2 = fig.add_subplot(122)
ax2.scatter(X_umap[:, 0], X_umap[:, 1], c=t, cmap="Spectral", s=5)
ax2.set_title("UMAP Projection (Unrolled Spring)")

plt.show()
```

## Introduction to Relations, Ordered Sets, and Lattices

### Relations

A **relation** is a fundamental concept in mathematics that describes a connection between elements of one or more sets. Formally, given two sets $ A $ and $ B $, a **binary relation** $ R $ from $ A $ to $ B $ is a subset of the Cartesian product $ A \times B $. That is,

$R \subseteq A \times B.$

If $ (a, b) \in R $, we say that $ a $ is related to $ b $ under $ R $, often denoted as $ a R b $.

### Properties of Relations
Relations may possess various properties, including:
- **Reflexivity**: $ \forall a \in A, (a, a) \in R $
- **Symmetry**: $ \forall a, b \in A, (a, b) \in R \Rightarrow (b, a) \in R $
- **Antisymmetry**: $ \forall a, b \in A, (a, b) \in R $ and $ (b, a) \in R \Rightarrow a = b $
- **Transitivity**: $ \forall a, b, c \in A, (a, b) \in R $ and $ (b, c) \in R \Rightarrow (a, c) \in R $

A relation that is reflexive, antisymmetric, and transitive is called a **partial order**. If also satisfies
- **comparability for all elements**: $ \forall a, b \in X, \quad a \leq b \quad \text{or} \quad b \leq a. $

it is called **total order** or **linear order**.

#### Special Elements in Ordered Sets
- **Greatest element**: An element $ g \in P $ such that $ \forall x \in P, x \leq g $.
- **Least element**: An element $ l \in P $ such that $ \forall x \in P, l \leq x $.
- **Maximal element**: An element $ m $ such that there is no $ x \in P $ with $ m < x $.
- **Minimal element**: An element $ n $ such that there is no $ x \in P $ with $ x < n $.

### Lattices
A **lattice** is a special type of poset in which every pair of elements $(a,b)$ has both a **least upper bound** $a \vee b$ (supremum, or **join**) and a **greatest lower bound** $a \wedge b$ (infimum, or **meet**). Formally, a poset $ (L, \leq) $ is a lattice if:


$\forall a, b \in L, \quad (a \vee b) \text{ and } (a \wedge b) \text{ exist in } L. $

#### Types of Lattices
- **Bounded Lattice**: A lattice with a greatest and least element (denoted as 1 and 0, respectively).
- **Distributive Lattice**: A lattice where join and meet operations distribute over each other.
- **Complemented Lattice**: A bounded lattice where every element has a complement.
- **Boolean Algebra**: A complemented distributive lattice, forming the basis for Boolean logic.

## Power Set

The **power set** $ \mathcal{P}(S)$  of a set $ S $ is the set of all possible subsets of $ S $, including the empty set $ \emptyset $ and $ S $ itself.

$\mathcal{P}(S) = \{ A \mid A \subseteq S \}$


If a set $ S $ has $ n $ elements, then the number of subsets (size of the power set) is:

$ |\mathcal{P}(S)| = 2^n $


### Examples
If \( S = \{a, b\} \), then the power set is:


$\mathcal{P}(S) = \{\emptyset, \{a\}, \{b\}, \{a, b\} \}$

## Formal Concept Analysis (FCA)

**Formal Concept Analysis (FCA)** is a mathematical framework for data analysis, classification, and knowledge representation based on lattice theory. It provides a method to extract and visualize conceptual structures from binary relations between objects and attributes.





A **formal context**  is a triplet $(G, M, R) $, where:
- $ G $ is the set of **objects**,
- $ M $ is the set of **attributes**,
- $ R \subseteq G \times M $ is the **relation** (indicating which objects have which attributes),

Order set example:
* $G=\{ER, WPO,SPO,WTO,STO,WQO,SQO\}$
* $M=\{r,s,a,t\}$
* And $R$ is given by the relation of the next table

| Name of the Set         | Reflexive (r) | Symmetric (s)  | Antisymmetric (a)  | Transitive (t) |
|-------------------------|-----------|-----------|--------------|------------|
| **Equivalence Relation** (ER) | X        |    X    |            | X         |
| **Weak Partial Order** (WPO)  | X        |         | X           | X         |
| **Strict Partial Order** (SPO) |         |         | X           | X         |
| **Weak Total Order** (WTO)    | X        |         | X           | X         |
| **Strict Total Order** (STO)  |         |         | X           | X         |
| **Weak Quasi-Order** (WQO)| X    |         |            | X         |
| **Strict Quasi-Order**   (SQO)   |         |         |            | X         |







### **Polar Maps**:

#### **Objects to Intent Map** ($ \overrightarrow{R} $)
The **Object to Intent Map**  assigns a set of objects $ A \subseteq G $ to the set of attributes they all share:


$\overrightarrow{R}(A) = \{ m \in M \mid \forall g \in A, (g, m) \in I \}$

This function returns all attributes that are common to every object in $ A $.

In the order set example

$\overrightarrow{R}(\{WTO,WQO\})=\{r,t\}$

#### **Attributes to Extent Map** ($ \overleftarrow{R} $)
The **Attribute to Extent Map** map assigns a set of attributes $ B \subseteq M $ to the set of objects that share all attributes in $ B $:

$\overleftarrow{R}(B) = \{ g \in G \mid \forall m \in B, (g, m) \in I \}$

This function returns all objects that possess every attribute in $ B $.

In the order set example

$\overleftarrow{R}(\{r,t\})=\{ER,WPO,WTO,WQO\}$

#### **Ext and Int Maps**
The map

$Ext(A)= \overleftarrow{R}(\overrightarrow{R}(A))$

Transform a subset of objects in a extent.  An the map

$Int(A)= \overrightarrow{R}(\overleftarrow{R}(B))$

Transform a subset of attributes in a intent.


### **Concept Lattice**  
A **formal concept** $ (A, B) $ is a pair where $ A = \overleftarrow{R}(B) $ and $ B = \overrightarrow{R}(A) $, meaning $ A $ and $ B $ are mutually closed

In a concept, $A$ is called an **Extent** and $B$ is an **Intent**.

Every object in $ A $ has all attributes in $ B $, and every attribute in $ B $ is shared by all objects in $ A $.


In the last example, the tuple $(\{ER,WPO,WTO,WQO\},\{r,t\})$ is a concept.

The set of all concepts ordered by Extent containment forms a lattice, and is called **concept lattice**

````python
from itertools import chain, combinations

class FormalContext:
    def __init__(self, objects, attributes, relation):
        """
        Initialize a formal context.
        :param objects: List of object names
        :param attributes: List of attribute names
        :param relation: List of (object, attribute) pairs defining the context
        """
        self.objects = objects
        self.attributes = attributes
        self.relation = {(o, a) for o, a in relation}

    def intent(self, obj_set):
        """Compute the intent (common attributes) of a given set of objects."""
        if not obj_set:
            return set(self.attributes)
        return {a for a in self.attributes if all((o, a) in self.relation for o in obj_set)}

    def extent(self, attr_set):
        """Compute the extent (common objects) of a given set of attributes."""
        if not attr_set:
            return set(self.objects)
        return {o for o in self.objects if all((o, a) in self.relation for a in attr_set)}

    def compute_concepts(self):
        """Compute all formal concepts using Next Closure Algorithm."""
        all_objects = set(self.objects)
        concepts = set()

        # Generate all subsets of objects
        for obj_subset in chain.from_iterable(combinations(all_objects, r) for r in range(len(all_objects)+1)):
            extent = set(obj_subset)
            intent = self.intent(extent)
            extent = self.extent(intent)  # Ensure closure
            concepts.add((frozenset(extent), frozenset(intent)))

        return sorted(concepts, key=lambda x: (len(x[0]), len(x[1])))

    def display_concepts(self):
        """Display all computed formal concepts."""
        concepts = self.compute_concepts()
        print("Formal Concepts:")
        for extent, intent in concepts:
            print(f"Extent: {set(extent)}, Intent: {set(intent)}")

# Example usage
objects = ['O1', 'O2', 'O3']
attributes = ['A1', 'A2', 'A3']
relation = [('O1', 'A1'), ('O1', 'A2'), ('O2', 'A2'), ('O2', 'A3'), ('O3', 'A1')]

context = FormalContext(objects, attributes, relation)
context.display_concepts()
````

## Association Rules

 The goal is to identify significant correlations between attributes within objects to support decision-making in knowledge discovery, classification, and data analysis.  


### **Attributes probabilities**

1. In a formal context $(G, M, R) $, the **probability of a subset of attributes** $ B \subseteq M $  is:

$P(B) = \frac{|\overleftarrow{R}(B)|}{|G|}$

Where:
- $ |\overleftarrow{R}(B)| $ is the number of objects that have all attributes in $ B $.
-  $ |G| $ is the total number of objects.



2. The **attributes conditional probability** of a subset $ B_2 \subseteq M $ given another subset $B_1 \subseteq M $  is defined as:

$$P(B_2 \mid B_1)  = \frac{P(B_1 \cup B_2)}{P(B_1)} = \frac{|\overleftarrow{R}(B_1 \cup B_2)|}{|\overleftarrow{R}(B_1)|}
$$



### **Association Rules**

In a formal context $(G, M, R) $, A **rule** $B_1 \Rightarrow  B_2$ is a pair of attributes subset (i. e. $B_1 \subseteq M$ and $B_2 \subseteq M$).  




1. The  **support of a subset of attributes** $ B \subseteq M $  is the probability of $B$:


$Sup(B) = P(B) = \frac{|\overleftarrow{R}(B)|}{|G|}$


2. The  **support of a rule** $B_1 \Rightarrow  B_2$  is the probability of $B_1 \cup B_2$:


$Sup(B_1 \Rightarrow  B_2) = P(B_1 \cup B_2) = \frac{|\overleftarrow{R}(B_1 \cup B_2)|}{|G|}$





3. The **confidence of a rule**  $B_1 \Rightarrow  B_2$ is defined as the attribute conditional probability:

$$Conf(B_1 \Rightarrow  B_2) = P(B_2 \mid B_1) = \frac{|\overleftarrow{R}(B_1 \cup B_2)|}{|\overleftarrow{R}(B_1)|}
$$


4. The **Lift of a rule**   $B_1 \Rightarrow B_2$ is defined as:  

$$ Lift(B_1\Rightarrow B_2) =\frac{Conf(B_1\Rightarrow B_2)}{Sup(B_2)} $$








### **Concept and rules**

A rule$ B_1 \Rightarrow B_2$​ has **100% confidence** if and only if:

$$B_2 \subseteq  \overrightarrow{R}(\overleftarrow{R}(B_1))$$

This means that all objects possessing $B_1$ must also possess $B_2$, making the rule deterministic rather than probabilistic.





````python
import polars as pl
from itertools import chain, combinations

class FormalContext:
    def __init__(self, objects, attributes, relation):
        """
        Initialize a formal context using Polars.
        :param objects: List of object names
        :param attributes: List of attribute names
        :param relation: List of (object, attribute) pairs defining the context
        """
        self.objects = objects
        self.attributes = attributes
        self.relation = pl.DataFrame({"object": [o for o, _ in relation], "attribute": [a for _, a in relation]})

    def intent(self, obj_set):
        """Compute the intent (common attributes) of a given set of objects."""
        if not obj_set:
            return set(self.attributes)
        filtered = self.relation.filter(pl.col("object").is_in(obj_set))
        return set(filtered.group_by("attribute").agg(pl.count()).filter(pl.col("count") == len(obj_set))["attribute"])

    def extent(self, attr_set):
        """Compute the extent (common objects) of a given set of attributes."""
        if not attr_set:
            return set(self.objects)
        filtered = self.relation.filter(pl.col("attribute").is_in(attr_set))
        return set(filtered.group_by("object").agg(pl.count()).filter(pl.col("count") == len(attr_set))["object"])

    def compute_concepts(self):
        """Compute all formal concepts using Next Closure Algorithm."""
        all_objects = set(self.objects)
        concepts = set()

        # Generate all subsets of objects
        for obj_subset in chain.from_iterable(combinations(all_objects, r) for r in range(len(all_objects)+1)):
            extent = set(obj_subset)
            intent = self.intent(extent)
            extent = self.extent(intent)  # Ensure closure
            concepts.add((frozenset(extent), frozenset(intent)))

        return sorted(concepts, key=lambda x: (len(x[0]), len(x[1])))

    def display_concepts(self):
        """Display all computed formal concepts."""
        concepts = self.compute_concepts()
        print("Formal Concepts:")
        for extent, intent in concepts:
            print(f"Extent: {set(extent)}, Intent: {set(intent)}")

class AssociationRules(FormalContext):
    def support(self, attr_set):
        """Compute the support of a subset of attributes."""
        extent_size = len(self.extent(attr_set))
        return extent_size / len(self.objects)

    def confidence(self, B1, B2):
        """Compute the confidence of a rule B1 => B2."""
        sup_B1_B2 = self.support(B1 | B2)
        sup_B1 = self.support(B1)
        return sup_B1_B2 / sup_B1 if sup_B1 > 0 else 0

    def lift(self, B1, B2):
        """Compute the lift of a rule B1 => B2."""
        conf_B1_B2 = self.confidence(B1, B2)
        sup_B2 = self.support(B2)
        return conf_B1_B2 / sup_B2 if sup_B2 > 0 else 0

    def generate_rules(self, min_conf=0.5, min_sup=0.1):
        """Generate association rules with minimum confidence and support."""
        rules = []
        attr_subsets = list(chain.from_iterable(combinations(set(self.attributes), r) for r in range(1, len(self.attributes))))

        for B1 in attr_subsets:
            for B2 in attr_subsets:
                if set(B1).isdisjoint(set(B2)):
                    B1, B2 = set(B1), set(B2)
                    sup = self.support(B1 | B2)
                    conf = self.confidence(B1, B2)
                    lift = self.lift(B1, B2)
                    if sup >= min_sup and conf >= min_conf:
                        rules.append((B1, B2, sup, conf, lift))

        return sorted(rules, key=lambda x: (-x[3], -x[2]))  # Sort by confidence, then support

    def display_rules(self, min_conf=0.5, min_sup=0.1):
        """Display generated association rules."""
        rules = self.generate_rules(min_conf, min_sup)
        print("Association Rules:")
        for B1, B2, sup, conf, lift in rules:
            print(f"Rule: {B1} => {B2}, Support: {sup:.3f}, Confidence: {conf:.3f}, Lift: {lift:.3f}")

# Example usage
objects = ['T1', 'T2', 'T3', 'T4']  # Transactions
attributes = ['Milk', 'Bread', 'Butter', 'Cheese']  # Items
relation = [('T1', 'Milk'), ('T1', 'Bread'), ('T2', 'Milk'), ('T2', 'Butter'), ('T3', 'Bread'), ('T3', 'Cheese'), ('T4', 'Milk'), ('T4', 'Bread'), ('T4', 'Butter')]

context = AssociationRules(objects, attributes, relation)
context.display_rules(min_conf=0.5, min_sup=0.25)
````

# Apriori Algorithm for Association Rule Mining

## Introduction


## How Apriori Works
The Apriori algorithm is based on the **Apriori principle**, which states:
> *If an itemset is frequent, then all of its subsets must also be frequent.*

The algorithm operates in two main steps:
1. **Frequent Itemset Generation**: Identify itemsets that satisfy a minimum support threshold.
2. **Association Rule Generation**: Extract strong association rules from the frequent itemsets that satisfy a minimum confidence threshold.

## Steps of the Apriori Algorithm
1. **Set Minimum Support and Confidence**
   - Define the thresholds for support and confidence.

2. **Generate Frequent Itemsets**
   - Count occurrences of individual items and remove those that do not meet the support threshold.
   - Generate candidate itemsets of size `k+1` from frequent itemsets of size `k`.
   - Prune itemsets that have infrequent subsets.
   - Repeat until no more frequent itemsets can be generated.

3. **Generate Association Rules**
   - For each frequent itemset, generate possible association rules.
   - Compute the confidence of each rule.
   - Keep rules that meet the minimum confidence threshold.

## Example
Consider a dataset of transactions:

| Transaction ID | Items Purchased          |
|---------------|-------------------------|
| 1             | Milk, Bread, Butter     |
| 2             | Milk, Bread             |
| 3             | Bread, Butter           |
| 4             | Milk, Butter            |
| 5             | Milk, Bread, Butter     |

### Step 1: Generate Frequent Itemsets
- **Support Calculation**:
  - `Milk`: 4/5 = 80%
  - `Bread`: 4/5 = 80%
  - `Butter`: 3/5 = 60%
  - `Milk, Bread`: 3/5 = 60%
  - `Milk, Butter`: 3/5 = 60%
  - `Bread, Butter`: 3/5 = 60%
  - `Milk, Bread, Butter`: 2/5 = 40%

Assuming `min_support = 50%`, we discard `Milk, Bread, Butter`.

### Step 2: Generate Association Rules
- From `{Milk, Bread} -> Butter`, calculate:
  - Confidence = (Support of `{Milk, Bread, Butter}`) / (Support of `{Milk, Bread}`) = 40% / 60% = 66.7%
  - If `min_confidence = 60%`, then keep this rule.

## Advantages
- **Efficient** for small and medium-sized datasets.
- **Simple and easy to implement**.

## Disadvantages
- **Computationally expensive** for large datasets.
- **Generates many candidate itemsets**, leading to high memory usage.

## Conclusion
The **Apriori algorithm** is a widely used method in market basket analysis and recommendation systems. However, due to its limitations, alternative algorithms like **FP-Growth** are often used for better performance in large datasets.



# FP-Growth Algorithm for Association Rule Mining


The **FP-Growth (Frequent Pattern Growth) algorithm** is an efficient method for mining frequent itemsets without generating candidate itemsets, unlike the Apriori algorithm. It uses a **compact data structure** called an **FP-Tree (Frequent Pattern Tree)** to improve performance.

## How FP-Growth Works

1. **Constructing an FP-Tree**: A compact representation of transactions is built by storing item frequencies and maintaining item order.
2. **Mining Frequent Patterns**: The tree is recursively traversed to generate frequent itemsets without candidate generation.

## Steps of the FP-Growth Algorithm
1. **Scan the dataset once** to count item frequencies.
2. **Sort items in descending order of frequency** and discard infrequent items.
3. **Build the FP-Tree**:
   - The root node is created as a null node.
   - Transactions are inserted as branches, sharing common prefixes.
4. **Mine the FP-Tree recursively** to generate frequent itemsets:
   - Identify conditional pattern bases.
   - Construct conditional FP-Trees.
   - Extract frequent itemsets.

## Example
Consider a dataset of transactions:

| Transaction ID | Items Purchased          |
|---------------|-------------------------|
| 1             | Milk, Bread, Butter     |
| 2             | Milk, Bread             |
| 3             | Bread, Butter           |
| 4             | Milk, Butter            |
| 5             | Milk, Bread, Butter     |

### Step 1: Calculate Item Frequency
```
Milk: 4, Bread: 4, Butter: 3
```

### Step 2: Construct the FP-Tree
```
Root → Milk(4) → Bread(3) → Butter(2)
              ↘ Butter(1)
```

### Step 3: Generate Frequent Itemsets
- `{Bread, Butter}` → 3 times
- `{Milk, Bread}` → 3 times
- `{Milk, Butter}` → 3 times
- `{Milk, Bread, Butter}` → 2 times

## Advantages
- **Faster** than Apriori for large datasets.
- **No candidate generation**, reducing computational cost.
- **Compact FP-Tree representation** improves efficiency.

## Disadvantages
- **Tree construction can be memory-intensive**.
- **Not intuitive to interpret manually**.

## Conclusion
The **FP-Growth algorithm** is a powerful alternative to Apriori, offering improved performance for association rule mining. It is widely used in market basket analysis, recommendation systems, and other data mining applications.
